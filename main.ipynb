```python
from sklearn.datasets import load_iris
X, y = load_iris(return_X_y=True, as_frame=True)
```

```python
import seaborn as sns
sns.heatmap(X.corr(numeric_only=True), annot=True)
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

#     
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.hist(X['sepal length (cm)'], bins=20)
plt.title('sepal length (cm)')

plt.subplot(2, 2, 2)
plt.hist(X['sepal width (cm)'], bins=20)
plt.title('sepal width (cm)')

plt.subplot(2, 2, 3)
plt.hist(X['petal length (cm)'], bins=20)
plt.title('petal length (cm)')

plt.subplot(2, 2, 4)
plt.hist(X['petal width (cm)'], bins=20)
plt.title('petal width (cm)')

plt.tight_layout()
plt.show()

#  qq-   
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
stats.probplot(X['sepal length (cm)'], dist='norm', plot=plt)
plt.title('sepal length (cm)')

plt.subplot(2, 2, 2)
stats.probplot(X['sepal width (cm)'], dist='norm', plot=plt)
plt.title('sepal width (cm)')

plt.subplot(2, 2, 3)
stats.probplot(X['petal length (cm)'], dist='norm', plot=plt)
plt.title('petal length (cm)')

plt.subplot(2, 2, 4)
stats.probplot(X['petal width (cm)'], dist='norm', plot=plt)
plt.title('petal width (cm)')

plt.tight_layout()
plt.show()
```

```python
X.sample(2)
```

```python
for i in X:
    #   
    plt.figure(figsize=(8, 6))
    plt.hist(X[i], bins=10, edgecolor='black', alpha=0.7, color='steelblue')
    plt.xlabel(i)
    plt.ylabel('')
    plt.title(f' {i}')
    plt.show()

    #     
    plt.figure(figsize=(8, 6))
    plt.boxplot(X[i], vert=True, patch_artist=True)
    plt.xlabel(i)
    plt.title(f' {i}')
    plt.show()
```

```python
import seaborn as sns
for i in X:
    sns.boxplot(X, y=i, x=y)
    plt.show()
```

```python

#    Iris
import pandas as pd


iris_data = pd.concat([X, y], axis=1)

#   
correlation_matrix = iris_data.corr()

#     
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('      Iris')
plt.show()

#        
correlation_pairs = correlation_matrix.unstack().sort_values(ascending=False)
strongest_correlation = correlation_pairs[correlation_pairs < 1].head(1)

#     
variable1, variable2 = strongest_correlation.index[0]

#         
plt.figure(figsize=(8, 6))
plt.scatter(iris_data[variable1], iris_data[variable2])
plt.xlabel(variable1)
plt.ylabel(variable2)
plt.title(f'  {variable1}  {variable2}')
plt.show()
```
