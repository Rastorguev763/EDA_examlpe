# ЗАДАНИЕ 1

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
dataset_0 = pd.read_csv('dataset_0.csv')
```

```python
dataset_0
```

```python

# Вычисление матрицы корреляции
correlation_matrix = dataset_0.corr()

# Построение тепловой карты матрицы корреляции
plt.figure(figsize=(15, 15))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Матрица корреляции переменных')
plt.show()
```

На матрице корреляции данных видно, что корреляция переменных **'feature_3', 'feature_4', 'feature_13', 'feature_17', 'feature_20'** близка к 0, это явно указывает на отсутсвие влияния этих перменных на остальные переменные в дата сете и **'feature_3', 'feature_4', 'feature_13', 'feature_17', 'feature_20'** случайный шум который нужно удалить из дата сета.

```python
columns_to_drop = ['feature_3', 'feature_4', 'feature_13', 'feature_17', 'feature_20']
dataset_0 = dataset_0.drop(columns=columns_to_drop)
dataset_0
```

# Задание 2

```python
dataset_1 = pd.read_csv('dataset_1.csv')
dataset_1.sample(5)
```

```python
dataset_1.info()
```

```python
# Удаление последнего столбца (целевого признака)
dataset_1_features = dataset_1.iloc[:, :-1]

for column in dataset_1_features.columns:
    
    # Диаграмма рассеяния между переменной X и таргетом
    plt.scatter(dataset_1[column], dataset_1['target'])
    plt.xlabel(column)
    plt.ylabel('Target')
    plt.title(f'Зависимость между {column} и Target')
    plt.show()

for column in dataset_1.columns:
    # Гистограмма для переменной Y
    plt.hist(dataset_1[column], bins=10)
    plt.xlabel(column)
    plt.ylabel('Частота')
    plt.title(f'Распределение {column}')
    plt.show()
```

```python
# Построение тепловой карты матрицы корреляции
plt.figure(figsize=(5, 5))
sns.heatmap(dataset_1.corr(), annot=True, cmap='coolwarm')
plt.title('Матрица корреляции переменных')
plt.show()
```

```python
import seaborn as sns
for i in dataset_1:
    sns.boxplot(dataset_1, y=i, x=dataset_1['target'])
    plt.show()
```

```python
from scipy import stats


for column in dataset_1.columns:
    # Построение гистограмм для каждой переменной
    plt.figure(figsize=(12, 6))
    
    plt.subplot(2, 2, 1)
    plt.hist(dataset_1[column], bins=20)
    plt.title(f'{column}')

    plt.tight_layout()
    plt.show()

    # Построение qq-графиков для каждой переменной
    plt.figure(figsize=(12, 6))

    plt.subplot(2, 2, 1)
    stats.probplot(dataset_1[column], dist='norm', plot=plt)
    plt.title(f'{column}')

    plt.tight_layout()
    plt.show()
```
